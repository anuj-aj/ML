```python
import numpy as np
```

```python
def run():
    data=np.loadtxt("C:\\Users\\hp\\data.csv",delimiter=",")
    learning_rate=0.0001
    num_iterations=1000
    m,c=gd(data,learning_rate,num_iterations)
    print(m,c)
```

```python
def gd(points,learning_rate,num_iterations):
    m=1
    c=1
    for i in range(num_iterations):
        m,c=step_grad(points,learning_rate,m,c)
        print(i, " Cost: ", cost(points, m, c))    
    return m,c   

```

```python
def step_grad(points,lr,m,c):
    
    m_slope=0
    c_slope=0
    n=len(points)
    for i in range(n):
        x=points[i,0]
        y=points[i,1]
        m_slope+=(-2/n)*(y-(m*x)-c)*(x)
        c_slope+=(-2/n)*(y-(m*x)-c)
    new_m = m-(lr*m_slope)
    new_c = c-(lr*c_slope)
    return new_m,new_c
```

```python
def cost(points, m, c):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i, 0]
        y = points[i, 1]
        total_cost += (1/M)*((y - m*x - c)**2)
    return total_cost
```

```python
run()
```

```python
data4=pd.read_sql_query("select Rating,Domestic,Worldwide from IMDB Join earning on IMDB.movie_id=earning.movie_id ",db)
data4
```

```python
import matplotlib.pyplot as plt


plt.scatter(data4.Rating,data4.Domestic)


plt.scatter(data4.Rating,data4.Worldwide)
plt.xlabel("Rating")
plt.ylabel("Earning")
plt.title("rating vs earning")
plt.show()
```

```python
data5=pd.read_sql_query("select Rating,Budget,Worldwide from IMDB Join earning on IMDB.movie_id=earning.movie_id ",db)
data5=data5.sort_values(by=['Worldwide'],ascending=False)
data5.head().Budget.mean()
```

```python
data6=pd.read_sql_query("select genre,Worldwide from earning Join genre on earning.movie_id=genre.movie_id ",db)
data6
```

```python
genreY=data6.groupby("genre").Worldwide.mean()
genreY
```

```python
import pandas as pd
import numpy as np
titanic = pd.read_csv("training_titanic_x_y_train.csv")
titanic.head()
```

```python
del titanic["Name"]
del titanic["Ticket"]
del titanic["Fare"]
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
del titanic["Cabin"]
titanic.head()
```

```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
titanic["gender"]=titanic["Sex"].apply(getNumber)
#We have created a new column called "gender" and filling it with values 1 ,2 based on the values of sex column
del titanic["Sex"]
#Deleted Sex column, since no use of it now
titanic.head()
```

```python
#Fill the null values of the Age column.
#Fill mean Survived age(mean age of the survived people) in the column where the person has survived
#and mean not Survived age (mean age of the people who have not survived) in the column where person has not survived

#finding mean survived age
meanS= titanic[titanic.Survived==1].Age.mean()
#Creating a new "age" column , filling values in it with a condition (given inside)
#if goes True then given values (here meanS) is put in place of last values
#else nothing happens, simply the values are copied from the "Age" column of the daatset
titanic["age"]=np.where(pd.isnull(titanic.Age) & titanic["Survived"]==1  ,meanS, titanic["Age"])

# Finding the mean age of "Not Survived" people
meanNS=titanic[titanic.Survived==0].Age.mean()
#Now the "age" column contains null entries in the places where survived is equal to zero. 
#Filling those null values in one go with meanNS
titanic.age.fillna(meanNS,inplace=True)
titanic.head()
```

```python
#We can safely delete the Age column
del titanic["Age"]

#Renaming "age" and "gender" columns
titanic.rename(columns={'age':'Age'}, inplace=True)
titanic.rename(columns={'gender':'Sex'}, inplace=True)
titanic.head()
```

```python
## We want to check if "Embarked" column is is important for analysis or not,
## that is whether survival of the person depends on the Embarked column value or not.

# Finding the number of people who have survived 
# given that they have embarked or boarded from a particular port
survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
print(survivedQ)
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 1].shape[0]
print(survivedS)
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 1].shape[0]
print(survivedC)
```

```python
# Finding the number of people who have not survived 
# given that they have embarked or boarded from a particular port
survivedNQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 0].shape[0]
print(survivedNQ)
survivedNS = titanic[titanic.Embarked == 'S'][titanic.Survived == 0].shape[0]
print(survivedNS)
survivedNC = titanic[titanic.Embarked == 'C'][titanic.Survived == 0].shape[0]
print(survivedNC)
```

```python
#As there are significant changes in the survival rate based on which port the passengers aboard the ship.
#We cannot delete the whole embarked column (It is useful). 
#One embarked is null, thus drop that NaN
titanic.dropna(inplace=True)
titanic.head()
```

```python
def getEmbarked(str):
    if str=="S":
        return 1
    elif str=="Q":
        return 2
    else:
        return 3
titanic["embarked"]=titanic["Embarked"].apply(getEmbarked)
del titanic["Embarked"]
titanic.rename(columns={'embarked':'Embarked'}, inplace=True)
titanic.head()
```
